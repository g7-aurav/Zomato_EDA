```python
import numpy as np
import pandas as pd
import seaborn as sb 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_data=pd.read_csv(r"C:\GAURAV\ExcelR\My Project\project 4 python\zomato bangalore data set\archive (1)\zomato.csv")
zomato_data.head()
```

```python
zomato_data.info()
```

```python
pd.isnull(zomato_data).sum()
```

```python
#deleting unnecessary columns
zomato=zomato_data.drop(['url','dish_liked','phone'],axis=1)
```

```python
#finding null
pd.isnull(zomato).sum()
```

```python
#total duplicat values
zomato.duplicated().sum()
```

```python
zomato.count()
```

```python
#removing duplicates
zomato.drop_duplicates(inplace=True)
```

```python
zomato.duplicated().sum()
```

```python
zomato.isnull().sum()
```

# Null values on heat map

```python
sns.heatmap(zomato.isnull())
```

```python
#removing Null value
zomato.dropna(how='any',inplace=True)
```

```python
zomato.isnull().sum()
```

```python
#summary of data
zomato.info()
```

```python
#reading column names
zomato.columns
```

```python
#changing column names
zomato=zomato.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
zomato.columns
```

```python
#checking cost format
zomato['cost'].unique()
```

```python
#converting cost datatype to string
zomato['cost']=zomato['cost'].astype(str)
#removing comma with nothing
zomato['cost']=zomato['cost'].apply(lambda x:x.replace(',',''))
#converting cost datatype to int
zomato['cost']=zomato['cost'].astype(int)
zomato.info()
```

```python
#rechecking the format
zomato['cost'].unique()
```

# cost v/s order type

```python
zomato.groupby('online_order').sum()[['cost']]
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x= "cost", hue="online_order", data=zomato)
plt.title('Relation Between cost and type of order')
plt.xticks(rotation=90)
plt.show()
```

```python
#checking rate type 
zomato['rate'].unique()
```

```python
#removing 'new' and '-' from rates
zomato=zomato.loc[zomato.rate!='NEW']
zomato=zomato.loc[zomato.rate!='-'].reset_index(drop=True)
```

```python
zomato.rate.count()
```

```python
zomato['rate'].unique()
```

```python
#removing '/5' from rates
remove_slash=lambda x: x.replace('/5','')
zomato.rate=zomato.rate.apply(remove_slash)
```

```python
zomato['rate'].unique()
```
